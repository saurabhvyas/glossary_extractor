```python
 # function to remove digits from string
def drop_digits(in_str):
    digit_list = "1234567890"
    for char in digit_list:
        in_str = in_str.replace(char, "")

    return in_str
```

```python
# function to remove a specific word from a string
# specifically , I am creating this for removing the word see as in OS ( see Operating Systems ) in vocabulary concepts
def remove_word(word,str): 

    stopwords = [word]
    querywords = str.split()
  #  print querywords

    resultwords  = [word for word in querywords if word.lower() not in stopwords]
    result = ' '.join(resultwords)

    return result
```

```python
def check_if_braces(str):
    if '(' in str and ')' in str and str.index(')') > str.index('('):
        return True
    else :
        return False
```

```python
# this function removes duplicates and removes words , only returns concept phrases
def filter_phrases(concepts):
    concepts_refined=list(set(concepts))  # remove duplicates , if any


  

    # count number of words in each phrase
    
    concepts2=[]
    for concept in concepts_refined:
        count = len(re.findall(r'\w+', concept))
    #print (count)
        if count != 1:
                concepts2.append(concept)
                
      # need to convert all concepts into lowercase
    concepts2=map(str.lower,concepts2)
    return concepts2  
```
